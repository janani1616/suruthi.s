```python
from bs4 import BeautifulSoup
```

```python
url=('https://www.flipkart.com/mobiles/pr?sid=tyy,4io&otracker=categorytree')
```

```python
response=requests.get(url)
```

```python
response
```

```python
soup=BeautifulSoup(response.text)
```

```python
soup
```

```python
mobiles=soup.select('._4rR01T')
```

```python
mobiles
```

```python
mobiles_list=[]
for item in mobiles:
    mobiles_list.append(item.text)
```

```python
mobiles_list
```

```python
ramrom
```
